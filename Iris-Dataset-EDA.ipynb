```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load from seaborn's built-in dataset
df = sns.load_dataset('iris')
```

```python
# Shape of the dataset
print("Shape:", df.shape)
```

```python
# Column names
print("Columns:", df.columns.tolist())
```

```python
# First few rows
print(df.head())
```

```python
# Basic info about dataset
print(df.info())
```

```python
# Descriptive statistics
print(df.describe())
```

```python
# Scatterpot
sns.pairplot(df, hue="species")
plt.suptitle("Scatter Plot of Iris Features", y=1.02)
plt.show()
```

```python
# Histogram

df.hist(figsize=(10, 8), edgecolor='black')
plt.suptitle("Histograms of Iris Features", y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Boxplot

plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Box Plot of Iris Features")
plt.show()
```

```python
for col in df.columns[:-1]:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='species', y=col, data=df)
    plt.title(f'Boxplot of {col} by Species')
    plt.show()
```
